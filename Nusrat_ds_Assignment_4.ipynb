**Perform Normalization Using Loop**

```python
import csv
```

```python
with open('/content/supershops.csv', 'r') as file:
    reader = csv.reader(file)
    data = list(reader)
```

```python
for row in data:
    for i in range(4):  # Assuming there are 4 string columns to normalize
        value = row[i].strip().lower()  
        row[i] = value
```

```python
with open('supershops_normalized.csv', 'w', newline='') as file:#The normalized data will be saved in a new file called 'supershops_normalized.csv'
    writer = csv.writer(file)
    writer.writerows(data)
```
